#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Load data (users.csv, posts.csv)

```python
users = pd.read_csv('users.csv')
users.head(10)
```

```python
posts = pd.read_csv('posts.csv')
posts.head(10)
```

#### 3. Rename id column to user_id

```python
users.rename(columns = {'id': 'user_id'}, inplace=True)
users.head(3)
```

#### 4. Rename id column to post_id and owner_user_id to user_id

```python
posts.rename(columns = {'id': 'post_id', 'owner_user_id': 'user_id'}, inplace=True)
posts.head(3)
```

#### 5. Define new dataframes for users and posts with the following selected columns:
    **users columns**: user_id, reputation,views,up_votes,down_votes
    **posts columns**: post_id, score,user_id,view_count,comment_count

```python
usersdf = users[['user_id', 'reputation', 'views', 'up_votes', 'down_votes']]
usersdf.head(3)
```

```python
postsdf = posts[['post_id', 'score','user_id','view_count','comment_count']]
postsdf.head(3)
```

#### 6. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_df = pd.merge(posts, users, on = 'user_id')
merged_df.head(5)
```

#### 7. How many missing values do you have in your merged dataframe? 

```python
# Sólo tenemos los valores para los cuales hay un 'user_id' en común en ambos dataframes

# A continuación extraemos una observación para un ejemplo de 'user_id' que está en ambos dfs
```

```python
posts[posts['user_id'] == 598427]
```

```python
users[users['user_id'] == 598427]
```

There are no missing values

#### Bonus: Identify extreme values in your merged dataframe, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder. Hint: post_id cannot have outliers!

```python
# Vamos sacando los histogramas para ver los posibles outliers
```

```python
merged_df[['score', 'view_count']].hist()
```

```python
# Interpretamos que no debería haber score negativos
(merged_df['score'] < 0).value_counts()
```

```python
# Guardamos los valores en un nuevo dataframe: dfneg

dfneg = merged_df[(merged_df['score'] < 0)]
dfneg.head(3)
```

```python
# Identificamos 16 outliers con 'views' > 5000
(merged_df['views'] > 5000).value_counts()
```

```python
merged_df[['up_votes', 'down_votes']].hist()
```

```python
merged_df['up_votes'].value_counts().plot()
```

```python
# Identificamos 10 outliers con 'up_votes' > 3000
(merged_df['up_votes'] > 3000).value_counts()
```

```python
# Guardamos los valores en un nuevo dataframe: dfupv

dfupv = merged_df[(merged_df['up_votes'] > 3000)]
dfupv.head(3)
```

```python
merged_df['down_votes'].value_counts().plot()
```

```python
# Identificamos 13 outliers con 'down_votes' > 800
(merged_df['down_votes'] > 800).value_counts()
```

```python
# Guardamos los valores en un nuevo dataframe: dfdov

dfdov = merged_df[(merged_df['down_votes'] > 800)]
dfdov.head(3)
```

```python
# Como no se lee bien el gráfico, hay una manera más fácil de leer la info de los años sin cambiar opciones del gráfico

(pd.to_datetime(merged_df['creation_date']).dt.year).value_counts()
```

```python
(pd.to_datetime(merged_df['last_access_date']).dt.year).value_counts()
```

```python
# Identificamos 19 outliers con 'last_access_date' < 2016

(pd.to_datetime(merged_df['last_access_date']).dt.year < 2016).value_counts()
```

```python
# Guardamos los valores en un nuevo dataframe: dfdate

dfdate = merged_df[(pd.to_datetime(merged_df['last_access_date']).dt.year < 2016)]
dfdate.head(3)
```

```python
# Ahora tenemos que juntar los dataframoes con los outliers: dfdate, dfdov, dfupv, dfview, dfrep, dfneg
```

```python
# Concatenamos todos los dfs, eliminamos duplicados y ordenamos por índice

outliers = pd.concat([dfdate, dfdov, dfupv, dfview, dfrep, dfneg]).drop_duplicates().sort_index()
outliers
```

```python
# Lo exportamos a un archivo dentro de esta carpeta

outliers.to_csv('outliers.csv')
```
